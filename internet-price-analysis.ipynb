**Import package**

```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
import matplotlib.pyplot as plt
import os
import plotly_express as px
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

# ***Explore data***

```python
df_user=pd.read_csv('/kaggle/input/1-gb-internet-price/worldwide internet users - users.csv')
df_speed=pd.read_csv('/kaggle/input/1-gb-internet-price/worldwide internet speed in 2022  - avg speed.csv')
df_price=pd.read_csv('/kaggle/input/1-gb-internet-price/worldwide internet prices in 2022 - IN 2022.csv')
```

**Top 10 countries with highest Internet speed**

```python
df_speed.rename(columns = {'Avg \n(Mbit/s)Ookla':'Speed(in Mb/s)'}, inplace = True)
df_speed_desc = df_speed.sort_values(by=['Speed(in Mb/s)'],ascending=False).head(10)
country, speed =  list(df_speed_desc['Country']), list(df_speed_desc['Speed(in Mb/s)'])
fig = plt.figure(figsize = (10, 10))

plt.bar(country, speed, color ='gray',width = 0.5)
#Setting the speed value in top of the bar chart
for i in range(len(country)):
        plt.text(i, speed[i], speed[i], ha = 'center')
plt.xticks(country, rotation=30, size=10)        
plt.xlabel("Country")
plt.ylabel("Speed in Mb/s")
plt.title("Internet Speed in various Countries")
plt.show()
```

**Top 10 countries with lowest Internet speed**

```python
df_speed_asc = df_speed.sort_values(by=['Speed(in Mb/s)'],ascending=True).head(10)
country, speed =  list(df_speed_asc['Country']), list(df_speed_asc['Speed(in Mb/s)'])
fig = plt.figure(figsize = (10, 10))

plt.bar(country, speed, color ='gray',width = 0.5)
#Setting the speed value in top of the bar chart
for i in range(len(country)):
        plt.text(i, speed[i], speed[i], ha = 'center')
plt.xticks(country, rotation=30, size=10)        
plt.xlabel("Country")
plt.ylabel("Speed in Mb/s")
plt.title("Internet Speed in various Countries")
plt.show()
```

**Top 10 countries with highest Internet prices**

```python
df_price.rename(columns = {'Name':'Country'}, inplace = True)
df_price.dtypes
df_price=df_price.loc[df_price['Average price of 1GB (USD)'].str.startswith('$', na=False)]
df_price['Average price of 1GB (USD)'] = df_price['Average price of 1GB (USD)'].str.replace('$', '')
df_price['Average price of 1GB (USD)'] = pd.to_numeric(df_price['Average price of 1GB (USD)'], downcast = 'float')
df_price

df_price_desc = df_price.sort_values(by=['Average price of 1GB (USD)'],ascending=False).head(10)
country, price =  list(df_price_desc['Country']), list(df_price_desc['Average price of 1GB (USD)'])
fig = plt.figure(figsize = (10, 10))

plt.bar(country, price, color ='blue',width = 0.5)
#Setting the speed value in top of the bar chart
for i in range(len(country)):
        plt.text(i, round(price[i],2), round(price[i],2), ha = 'center')
plt.xticks(country, rotation=60, size=12)        
plt.xlabel("Country")
plt.ylabel("Average price of 1GB (USD)")
plt.title("Internet Prices in various Countries")
plt.show()
```

**Top 10 countries with lowest Internet prices**

```python
df_price_asce = df_price.sort_values(by=['Average price of 1GB (USD)'],ascending=True).head(10)
country, price =  list(df_price_asce['Country']), list(df_price_asce['Average price of 1GB (USD)'])
fig = plt.figure(figsize = (10, 10))

plt.bar(country, price, color ='blue',width = 0.5)
#Setting the speed value in top of the bar chart
for i in range(len(country)):
        plt.text(i, round(price[i],2), round(price[i],2), ha = 'center')
plt.xticks(country, rotation=90, size=12)        
plt.xlabel("Country")
plt.ylabel("Average price of 1GB (USD)")
plt.title("Internet Prices in various Countries")
plt.show()
```

**Percentage of Internet Users according to region**

```python
df_user['Internet users'] = df_user['Internet users'].str.replace(',', '', regex=False)
df_user['Population'] = df_user['Population'].str.replace(',', '', regex=False)
df_user['Internet users']=df_user['Internet users'].astype(float)
df_user['Population']=df_user['Population'].astype(float)
df_user_sum = df_user.groupby('Region')['Internet users'].sum()
region = list(df_user_sum.index)
users = list(df_user_sum.values)
fig1, ax1 = plt.subplots(figsize=(10, 8))
ax1.pie(users,explode = [0,0,0,0,0.5], labels=region, autopct='%1.1f%%', shadow=False, startangle=90)
ax1.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.title('Percentage of Internet Users according to region')
plt.show()
```

**Merge three file**

```python
df=pd.merge(df_price, df_user, left_on='Country', right_on="Country or area", how='left')
df=pd.merge(df, df_speed, left_on='Country', right_on="Country", how='left')
df=df.drop(['Continental region', 'Country or area','Country'], axis=1)
df.dropna()
```

**Compare Internet user percentage each region with Internet price**

```python
df['Internet user percentage']=df['Internet users']/df['Population']*100
Percentage = df.groupby('Region').sum()['Internet user percentage']
prices = df.groupby('Region').mean()['Average price of 1GB (USD)']
per = [Percentage for Percentage, quant in Percentage.items()]

x = per
y1 = Percentage
y2 = prices

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(x, y1, color='g')
ax2.plot(x, y2, color='b')

ax1.set_xticklabels(per, rotation=0, size=12)
ax1.set_xlabel('Region')
ax1.set_ylabel('Internet user percentage', color='g')
ax2.set_ylabel('Average price of 1GB (USD)', color='b')

plt.show()
```
